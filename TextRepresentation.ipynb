# Problem 1

# Apply all the preprocessing techniques that you think are necessary

```python
df['review'] = df['review'].str.lower()
```

```python
def remove_html_tags(text):
    return re.sub(r'<[^>]+>', '', text)

df['review'] = df['review'].apply(remove_html_tags)
```

```python
import string

exclude = string.punctuation

def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))

df['review'] = df['review'].apply(remove_punc)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Load stopwords once
stop_words = set(stopwords.words('english'))

def remove_stop_words(text):
    if not isinstance(text, str):
        return text
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)

df['review'] = df['review'].apply(remove_stop_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True)
x = cv.fit_transform(df['review'])

```

```python
# Problem 2

# Find out the number of words in the entire corpus and also the total number of unique words(vocabulary) using just python

total_words = 0
vocab = set()

for text in df['review']:
    
    text = str(text).lower()      
    words = text.split()          
    
    total_words += len(words)
    
    for word in words:
        vocab.add(word)

print("Total number of words:", total_words)
print("Total unique words (vocabulary size):", len(vocab))

```
